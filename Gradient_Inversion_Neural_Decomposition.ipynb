```python
import torch
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import random_split, DataLoader

torch.manual_seed(42)


transfrom = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])


full_dataset = torchvision.datasets.CIFAR10(
    root='./data',
    train=True,
    download=True,
    transform=transfrom
)


total_size = len(full_dataset)
train_size = int(0.65 * total_size)
val_size   = int(0.25 * total_size)
reserved_size = total_size - train_size - val_size

print(f"Dataset split: Train={train_size}, Validation={val_size}, Reserved={reserved_size}")


generator = torch.Generator().manual_seed(42)

train_dataset, val_dataset, reserved_dataset = random_split(
    full_dataset,
    [train_size, val_size, reserved_size],
    generator=generator
)


train_loader = DataLoader(train_dataset, batch_size=128, shuffle=True, num_workers=0, pin_memory=True)
val_loader   = DataLoader(val_dataset,   batch_size=128, shuffle=False, num_workers=0, pin_memory=True)
reserved_loader = DataLoader(reserved_dataset, batch_size=128, shuffle=False, num_workers=0, pin_memory=True)

for images, labels in train_loader:
    print(f"Train batch shape: {images.shape}, labels: {labels.shape}")
    break

print("✅ CIFAR-10 successfully split into train (65%), validation (25%), and reserved (10%) sets!")
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
from tqdm import tqdm


class SimpleCIFARNet(nn.Module):
    def __init__(self, num_classes=10):
        super().__init__()

        self.features = nn.Sequential(
            nn.Conv2d(3, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(inplace=True),
            nn.Conv2d(32, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2, 2),

            # Block 2
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.Conv2d(64, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2, 2),

            # Block 3
            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2, 2)
        )

        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(128 * 4 * 4, 512),
            nn.ReLU(inplace=True),
            nn.Dropout(0.5),
            nn.Linear(512, num_classes)
        )

    def forward(self, x):
        x = self.features(x)
        x = self.classifier(x)
        return x


def train_model(model, train_loader, val_loader, epochs=30, lr=0.001, device='cuda'):
    model = model.to(device)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=lr, weight_decay=1e-4)
    scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=10, gamma=0.5)

    train_losses, val_losses, val_accs = [], [], []

    for epoch in range(epochs):
        model.train()
        running_loss = 0.0
        correct = 0
        total = 0

        for images, labels in tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}"):
            images, labels = images.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

        train_loss = running_loss / len(train_loader)
        train_acc = 100. * correct / total
        train_losses.append(train_loss)

        # Validation
        model.eval()
        val_loss = 0.0
        correct = 0
        total = 0

        with torch.no_grad():
            for images, labels in val_loader:
                images, labels = images.to(device), labels.to(device)
                outputs = model(images)
                loss = criterion(outputs, labels)
                val_loss += loss.item()
                _, predicted = torch.max(outputs, 1)
                total += labels.size(0)
                correct += (predicted == labels).sum().item()

        val_loss /= len(val_loader)
        val_acc = 100. * correct / total
        val_losses.append(val_loss)
        val_accs.append(val_acc)

        scheduler.step()

        print(f"Epoch {epoch+1:2d} | Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.2f}% | "
              f"Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.2f}%")

    # Quick plot
    plt.figure(figsize=(10,4))
    plt.subplot(1,2,1); plt.plot(train_losses, label='train'); plt.plot(val_losses, label='val'); plt.legend(); plt.title('Loss')
    plt.subplot(1,2,2); plt.plot(val_accs); plt.title('Validation Accuracy'); plt.ylabel('%')
    plt.tight_layout(); plt.show()

    return model


train_transform = transforms.Compose([
    transforms.Normalize((0.4914, 0.4822, 0.4465), (0.2470, 0.2435, 0.2616))
])


test_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.4914, 0.4822, 0.4465), (0.2470, 0.2435, 0.2616))
])

from torchvision.datasets import CIFAR10
from torch.utils.data import random_split

full = CIFAR10(root='./data', train=True, download=True, transform=None)

total = len(full)
train_size = int(0.65 * total)
val_size   = int(0.25 * total)
res_size   = total - train_size - val_size

train_ds, val_ds, res_ds = random_split(full, [train_size, val_size, res_size],
                                         generator=torch.Generator().manual_seed(42))

train_ds.dataset.transform = train_transform
val_ds.dataset.transform   = test_transform
res_ds.dataset.transform   = test_transform

# DataLoaders: set num_workers to 0 to avoid the _utils error
BATCH_SIZE = 128
train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True,  num_workers=0, pin_memory=True)
val_loader   = DataLoader(val_ds, batch_size=BATCH_SIZE*2, shuffle=False, num_workers=0, pin_memory=True)
res_loader   = DataLoader(res_ds, batch_size=BATCH_SIZE*2, shuffle=False, num_workers=0, pin_memory=True)

device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Using device: {device}")

model = SimpleCIFARNet()
trained_model = train_model(model, train_loader, val_loader, epochs=25, lr=0.001, device=device)
```

```python
import torch
import torch.nn as nn
import copy
from collections import defaultdict

# ────────────────────────────────────────────────
#   ASSUMING you already ran the previous cell:
#   - model, train_loader, device, criterion are ready
#   - optimizer is Adam (we'll use its lr for the calculation)
# ────────────────────────────────────────────────

model.to(device)
model.train()
optimizer.zero_grad()                     # clean state

# Take exactly ONE batch (as in previous experiment)
print("Taking the first batch from train_loader...")
images, labels = next(iter(train_loader))
images = images.to(device)
labels = labels.to(device)

# ── 1. Record weights BEFORE update ───────────────────────────────
before_params = {name: param.clone().detach()
                 for name, param in model.named_parameters()}

# ── 2. Forward + Backward (this populates .grad) ───────────────────
outputs = model(images)
loss = criterion(outputs, labels)
loss.backward()

# ── 3. Save the TRUE original gradients (right now, before any step)
# This is what .backward() actually computed
true_grad_dict = {}
true_grad_vector_parts = []
for name, param in model.named_parameters():
    if param.grad is not None:
        g = param.grad.clone().detach()
        true_grad_dict[name] = g
        true_grad_vector_parts.append(g.view(-1))

true_grad_vector = torch.cat(true_grad_vector_parts)
print(f"True gradient vector shape: {true_grad_vector.shape} "
      f"({true_grad_vector.numel():,} parameters)")

# ── 4. Perform the optimizer step (weights now change) ─────────────
optimizer.step()

# ── 5. Record weights AFTER update ────────────────────────────────
after_params = {name: param.clone().detach()
                for name, param in model.named_parameters()}

# ── 6. Calculate "original gradients" using your requested equation
# Δw = after - before
# gradient ≈ - Δw / lr     (this recovers the direction .backward() used)
lr = optimizer.param_groups[0]['lr']          # current learning rate (0.001 by default)

approx_grad_dict = {}
approx_grad_vector_parts = []

for name in before_params:
    delta = after_params[name] - before_params[name]          # change in weights
    approx_grad = - (delta / lr)                              # ← YOUR EQUATION
    approx_grad_dict[name] = approx_grad
    approx_grad_vector_parts.append(approx_grad.view(-1))

approx_grad_vector = torch.cat(approx_grad_vector_parts)

print(f"\nApproximate gradient vector (from Δw / lr) shape: {approx_grad_vector.shape}")
print(f"Approximate gradient L2 norm: {torch.norm(approx_grad_vector).item():.6f}")

# ── 7. Save both vectors for later use (e.g. analysis, cosine similarity, etc.)
torch.save(true_grad_vector,   'true_gradient_vector.pt')
torch.save(approx_grad_vector, 'approx_gradient_vector_from_weight_change.pt')

print("\n✅ Saved:")
print("   • true_gradient_vector.pt")
print("   • approx_gradient_vector_from_weight_change.pt")

# ── 8. Quick sanity check (difference should be small with plain SGD,
#       larger with Adam because of momentum/adaptive terms)
diff_norm = torch.norm(approx_grad_vector - true_grad_vector).item()
print(f"\nL2 difference between true grad and approx grad: {diff_norm:.8f}")
if diff_norm < 1e-6:
    print("   → Perfect match! (This happens with plain SGD)")
else:
    print("   → Expected difference because we are using Adam "
          "(not plain SGD). The approximation is still very useful.")

# Optional: quick per-layer comparison
print("\nTop 5 layers by gradient norm (approx):")
sorted_layers = sorted(approx_grad_dict.items(),
                       key=lambda x: torch.norm(x[1]).item(), reverse=True)
for name, g in sorted_layers[:5]:
    print(f"  {name: <50} norm = {torch.norm(g).item():.6f}")
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from tqdm import tqdm

# Helper: flatten all gradients into a single vector
def get_flat_grad(model):
    grads = []
    for p in model.parameters():
        if p.grad is not None:
            grads.append(p.grad.view(-1))
    return torch.cat(grads)

# ── 1. Create the "decomposition network" ─────────────────────────────
# This is a simple linear layer: 10 class fractions → full gradient vector
# Its 10 columns (after .T) will be the 10 learned independent component vectors

num_params = None
# Run one dummy batch just to know the gradient size
model.zero_grad()
images, labels = next(iter(train_loader))
images, labels = images.to(device), labels.to(device)
outputs = model(images)
loss = criterion(outputs, labels)
loss.backward()
flat_grad = get_flat_grad(model)
num_params = flat_grad.numel()
model.zero_grad()                     # clean up

print(f"Gradient vector size: {num_params:,} parameters")

decomp_net = nn.Linear(10, num_params, bias=False).to(device)
decomp_optimizer = optim.Adam(decomp_net.parameters(), lr=0.01)

# ── 2. Training loop for the decomposition ─────────────────────────────
# We sample many batches, compute the true gradient, and teach decomp_net
# that:   observed_grad  ≈  class_fractions @ component_matrix

NUM_ITERATIONS = 2000
ORTHO_LAMBDA = 1.0

decomp_net.train()
model.eval()                  # we only want gradients, no training of main model

print("Training gradient decomposer (10 independent class components)...")

for it in tqdm(range(NUM_ITERATIONS)):
    images, labels = next(iter(train_loader))      # fresh random batch each time
    images, labels = images.to(device), labels.to(device)

    model.zero_grad()
    outputs = model(images)
    loss = criterion(outputs, labels)
    loss.backward()

    observed_grad = get_flat_grad(model)           # shape (P,)
    model.zero_grad()                              # clean for next iteration

    # class fractions (what the user asked: 10 components for 10 classes)
    class_counts = torch.bincount(labels, minlength=10).float()
    class_frac = (class_counts / labels.size(0)).unsqueeze(0).to(device)   # (1, 10)

    # forward through decomposer
    predicted_grad = decomp_net(class_frac).squeeze(0)   # (P,)

    # reconstruction loss
    recon_loss = F.mse_loss(predicted_grad, observed_grad)

    # independence / orthogonality loss (makes the 10 components as independent as possible)
    components = decomp_net.weight                     # (P, 10)
    gram = components.T @ components                   # (10, 10)
    ortho_loss = (gram - torch.eye(10, device=device)).pow(2).mean()

    total_loss = recon_loss + ORTHO_LAMBDA * ortho_loss

    decomp_optimizer.zero_grad()
    total_loss.backward()
    decomp_optimizer.step()

    if (it + 1) % 50 == 0:
        recon = recon_loss.item()
        ortho = ortho_loss.item()
        print(f"  Iter {it+1:3d} | Recon MSE: {recon:.4e} | Ortho loss: {ortho:.4e}")

# ── 3. Extract the 10 independent component vectors ─────────────────────
# Each is a full-sized gradient-direction vector for one class
components = decomp_net.weight.T                   # shape (10, num_params)
torch.save(components, 'class_independent_components.pt')

print("\n✅ Done! Saved 10 independent components → 'class_independent_components.pt'")
print(f"Each component is a {num_params:,}-dimensional vector (one per class)")

# Quick check: how orthogonal are they now?
gram = (components @ components.T)
print(f"Off-diagonal Gram matrix mean: {(gram - torch.eye(10, device=device)).abs().mean().item():.6f}")
diag_norms = torch.diag(gram.detach()).cpu().numpy().round(4)
print("Diagonal norms (should be similar):", diag_norms)
```

```python
import torch
import torch.nn as nn
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
from tqdm.auto import tqdm

# ── Same lightweight LeNet ──────────────────────────────────────────────────
class LeNet(nn.Module):
    def __init__(self, channel=3, hidden=768, num_classes=10):
        super().__init__()
        act = nn.ReLU
        self.body = nn.Sequential(
            nn.Conv2d(channel, 12, kernel_size=5, padding=2, stride=2), act(),
            nn.Conv2d(12, 12, kernel_size=5, padding=2, stride=2),      act(),
            nn.Conv2d(12, 12, kernel_size=5, padding=2, stride=1),      act(),
        )
        self.fc = nn.Sequential(nn.Linear(hidden, num_classes))

    def forward(self, x):
        return self.fc(self.body(x).view(x.size(0), -1))

def weights_init(m):
    if hasattr(m, "weight"): m.weight.data.uniform_(-0.5, 0.5)
    if hasattr(m, "bias"):   m.bias.data.uniform_(-0.5, 0.5)

# ── Setup ───────────────────────────────────────────────────────────────────
device = 'cuda' if torch.cuda.is_available() else 'cpu'
criterion = nn.CrossEntropyLoss().to(device)
tp = transforms.ToPILImage()
class_names = ['airplane','automobile','bird','cat','deer','dog','frog','horse','ship','truck']

inv_net = LeNet().to(device)
inv_net.apply(weights_init)
inv_net.eval()

# ── TV loss (slightly higher for stability) ─────────────────────────────────
def total_variation_loss(img, tv_weight=3e-3):
    bs, _, h, w = img.shape
    tv_h = (img[:, :, 1:, :] - img[:, :, :-1, :]).pow(2).mean()
    tv_w = (img[:, :, :, 1:] - img[:, :, :, :-1]).pow(2).mean()
    return tv_weight * (tv_h + tv_w)

# ── Parameters (change these) ───────────────────────────────────────────────
BATCH_SIZE = 4                  # number of real images whose gradients are averaged
ITERATIONS = 350
TV_WEIGHT  = 3e-3

# ── Collect MEAN gradients from BATCH_SIZE images per class ─────────────────
print(f"Collecting MEAN gradients | batch_size={BATCH_SIZE}")

prior_data  = {}
prior_grads = {}

for cls in range(10):
    accumulated_grads = None
    count = 0
    gt_images = []

    for imgs, lbls in res_loader:
        mask = (lbls == cls)
        if mask.any():
            selected = imgs[mask][:BATCH_SIZE - count].to(device)
            gt_images.append(selected)
            count += selected.size(0)

            out = inv_net(selected)
            y = criterion(out, lbls[mask][:selected.size(0)].to(device))
            dy_dx = torch.autograd.grad(y, inv_net.parameters())

            if accumulated_grads is None:
                accumulated_grads = [g.detach().clone() for g in dy_dx]
            else:
                for i in range(len(accumulated_grads)):
                    accumulated_grads[i] += dy_dx[i].detach()

            if count >= BATCH_SIZE:
                break

    # Average the gradients
    for i in range(len(accumulated_grads)):
        accumulated_grads[i] /= count

    prior_grads[cls] = accumulated_grads
    prior_data[cls]  = torch.cat(gt_images, dim=0)   # keep originals for display

print("✅ Priors collected\n")

# ── Reconstruction (single dummy per class) ─────────────────────────────────
fig, axes = plt.subplots(2, 10, figsize=(24, 5.5))

cifar_mean = torch.tensor([0.4914, 0.4822, 0.4465]).view(3, 1, 1)
cifar_std  = torch.tensor([0.2470, 0.2435, 0.2616]).view(3, 1, 1)

for cls in range(10):
    original_dy_dx = prior_grads[cls]

    label_pred = torch.argmin(
        torch.sum(original_dy_dx[-2], dim=-1), dim=-1
    ).detach().reshape((1,)).to(device)

    dummy_data = torch.randn(1, 3, 32, 32, device=device, requires_grad=True)
    optimizer = torch.optim.LBFGS([dummy_data], lr=1.0)

    pbar = tqdm(range(ITERATIONS), leave=False, desc=class_names[cls])

    for it in pbar:
        def closure():
            optimizer.zero_grad()
            dummy_loss = criterion(inv_net(dummy_data), label_pred)
            dummy_dy_dx = torch.autograd.grad(dummy_loss, inv_net.parameters(), create_graph=True)

            grad_diff = sum(((gx - gy) ** 2).sum() for gx, gy in zip(dummy_dy_dx, original_dy_dx))
            tv = total_variation_loss(dummy_data, tv_weight=TV_WEIGHT)

            total_loss = grad_diff + tv
            total_loss.backward()
            return total_loss

        optimizer.step(closure)
        loss = closure().item()
        pbar.set_postfix(loss=f"{loss:.2e}")
        if loss < 1e-6:
            break

    # Plot
    orig_display = (prior_data[cls][0].cpu() * cifar_std + cifar_mean).clamp(0, 1)
    axes[0, cls].imshow(tp(orig_display))
    axes[0, cls].set_title(class_names[cls], fontsize=9)
    axes[0, cls].axis('off')

    recon_display = (dummy_data[0].cpu() * cifar_std + cifar_mean).clamp(0, 1)
    axes[1, cls].imshow(tp(recon_display))
    axes[1, cls].set_title(f"loss={loss:.1e}", fontsize=8)
    axes[1, cls].axis('off')

axes[0, 0].set_ylabel("Original (example)", fontsize=11)
axes[1, 0].set_ylabel("Reconstructed (from\nmean gradient)", fontsize=11)

plt.suptitle(f"iDLG — Mean-Gradient Reconstruction (batch_size={BATCH_SIZE}, ReLU + TV Loss)", fontsize=13)
plt.tight_layout()
plt.show()
print("✅ Done — this version converges reliably")
```
